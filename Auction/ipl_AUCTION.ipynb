```python
import seaborn as sns
players = df['Player'].unique()
unique_years = sorted(df['Year'].unique())
```

```python
player_counts = df['Player'].value_counts()
players_to_plot = player_counts[player_counts > 2].index
```

```python
i=0
for player in players_to_plot:
    player_data = df[df['Player'] == player]
    plt.figure(figsize=(10, 6))

    # Scatter plot with year on x-axis, Amount on y-axis, color by Team
    sns.scatterplot(
        data=player_data,
        x='Year',
        y='Amount',
        hue='Team',
        palette='Set2',
        s=200
    )

    # Customize the plot
    plt.title(f'Auction Price and Team for {player} Over the Years', fontsize=16)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Auction Price (Amount)', fontsize=12)
    plt.xticks(unique_years)  # Ensure only actual years are displayed
    plt.grid(True)
    plt.legend(title='Team', fontsize=10)
    plt.tight_layout()
    plt.show()
    i+=1
    if i==15:
      break
```
